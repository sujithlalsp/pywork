```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Circle

lat7 = np.genfromtxt('/home/sps/Downloads/anju_track_f/anju_track/lat_radiosonde.txt', delimiter=",")
lon7 = np.genfromtxt('/home/sps/Downloads/anju_track_f/anju_track/lon_radiosonde.txt', delimiter=",")

fig = plt.figure(figsize=(16, 16))
m = Basemap(llcrnrlon=74.5, llcrnrlat=8.5, urcrnrlon=78.5, urcrnrlat=11.5, projection='merc', resolution='h')

m.drawcoastlines()
m.shadedrelief()
m.drawparallels(np.arange(-90., 90., .1), labels=[1, 0, 0, 0], fontsize=16, linewidth=0.5)
m.drawmeridians(np.arange(-180., 180., .1), labels=[0, 0, 0, 1], fontsize=16, linewidth=0.5)

# Coordinates of the station
station_lon, station_lat = [76.332, 10.04229]

# Plot station marker with a yellow-filled circle
x_station, y_station = m(station_lon, station_lat)
m.plot(x_station, y_station, 'yo', markersize=10, label='Station')

# Plot circles at intervals of 10 km around the station from 10 km to 50 km
for radius in range(10, 51, 10):
    # Convert radius from kilometers to degrees
    circle_radius_deg = radius / 111.32

    # Create a Circle patch
    circle = Circle((station_lon, station_lat), radius=circle_radius_deg, fill=False, color='k', linewidth=1)

    # Add the circle to the plot
    plt.gca().add_patch(circle)

# Plot the radiosonde track
x7, y7 = m(lon7, lat7)
plt.plot(x7, y7, 'maroon', linestyle='-', linewidth=4, markersize=4, label='Radiosonde Track')

# Add legend
plt.legend(loc='upper left', shadow=True, fontsize='16')

# Add a title
plt.title('Radiosonde Track with Station and Circles on Basemap (Mercator Projection)')

# Save the plot or display it
plt.savefig('/home/sps/Desktop/track_with_station.jpg')
plt.show()
```
